```python
# load big datasets
file_c = '/fs/nexus-scratch/rhaworth/deep-conus/current/maxvals.csv'
file_f = '/fs/nexus-scratch/rhaworth/deep-conus/future/maxvals.csv'

curr_full = pd.read_csv(file_c)
fut_full = pd.read_csv(file_f)

varlist =  ['tk_1km', 'tk_3km', 'tk_5km', 'tk_7km',
            'ev_1km', 'ev_3km', 'ev_5km', 'ev_7km',
            'eu_1km', 'eu_3km', 'eu_5km', 'eu_7km', 
            'qv_1km', 'qv_3km', 'qv_5km', 'qv_7km',
            'pr_1km', 'pr_3km', 'pr_5km', 'pr_7km',
            'wmax', 'dbz', 'ctt']
```

```python
# combined plot for single var in big dataset
varnum = 0 # choose from list

sns.histplot(data=curr_full, x=varlist[varnum], bins=25)
sns.histplot(data=fut_full, x=varlist[varnum], bins=25)
plt.legend(['Current', 'Future'])
plt.yscale('log')
plt.title('Current vs Future Storms Max ' + varlist[varnum])
plt.show()
```

```python
# generate all plots
for var in varlist:
    sns.histplot(data=curr_full, x=var, bins=25)
    sns.histplot(data=fut_full, x=var, bins=25)
    plt.legend(['Current', 'Future'])
    plt.yscale('log')
    plt.title('Current vs Future Storms Max ' + var)
    plt.savefig('hist-' + var + '-compared.png', dpi=300)
    plt.clf()
```
